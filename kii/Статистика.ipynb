```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
labels = ['К1', 'К2', 'К3', 'К4']
s2019 = [347, 357, 379, 425]
s2020 = [521, 567, 583, 600]
s2021 = [607, 609, 580, 620]
s2022 = [714, 702, 774, 731]

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

plt.rc('font', **font)

x = np.arange(len(labels))  
width = 0.2

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, s2019, width, label='2019')
rects2 = ax.bar(x, s2020, width, label='2020')
rects3 = ax.bar(x + width, s2021, width, label='2021')
rects4 = ax.bar(x + 2 * width, s2022, width, label='2022')

ax.set_title('Количество инцидентов по кварталам')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.set_size_inches(16, 7, forward=True)

fig.tight_layout()

plt.show()
```

```python
labels = ['Утечка информации', 'Нарушение деятельности', 'Ущерб интересам государства', 'Финансовые потери', 
          'Использование ресурсов для проведения атак', 'Другое', 'Неизвестно']
df = pd.DataFrame(dict(graph=labels,
                  n=[51, 36, 7, 6, 5, 1, 22], 
                  m=[67, 4, 2, 24, 3, 4, 12])) 
ind = np.arange(len(df))
width = 0.4

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

plt.rc('font', **font)

fig, ax = plt.subplots()
ax.barh(ind, df.n, width, color='red', label='Организации')
ax.barh(ind + width, df.m, width, color='green', label='Частные лица')

ax.set(yticks=ind + width, yticklabels=df.graph, ylim=[2*width - 1, len(df)])
ax.set_xlim([0, 100])
ax.legend()
ax.invert_yaxis()
fig.set_size_inches(16, 7, forward=True)

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(aspect="equal"))

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

data = [44, 29, 17, 3, 7]
ingredients = ['Учётные данные','Персональные данные', 'Данные платёжных карт', 'Переписка', 'Другое']


def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=16, weight="normal")

ax.set_title("19% атак направлены против частных лиц")

fig.set_size_inches(13, 8, forward=True)

plt.show()
```
