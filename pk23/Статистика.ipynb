```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ['1', '2', '3', '4']
s2019 = [347, 357, 379, 425]
s2020 = [521, 567, 583, 600]
s2021 = [607, 609, 580, 620]
s2022 = [714, 702, 774, 731]
s2023 = [786, 821, 0, 0]

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

plt.rc('font', **font)

x = np.arange(len(labels))  
width = 0.15

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2 * width, s2019, width, label='2019')
rects2 = ax.bar(x - width, s2020, width, label='2020')
rects3 = ax.bar(x, s2021, width, label='2021')
rects4 = ax.bar(x + width, s2022, width, label='2022')
rects5 = ax.bar(x + 2 * width, s2023, width, label='2023')

ax.set_title('   ')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.set_size_inches(16, 7, forward=True)

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(aspect="equal"))

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

data = [43, 26, 18, 3, 10]
ingredients = [' ',' ', '  ', '', '']


def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=16, weight="normal")

ax.set_title("15%     ")

fig.set_size_inches(13, 8, forward=True)

plt.show()
```
