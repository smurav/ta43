```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ['К1', 'К2', 'К3', 'К4']
s2020 = [521, 567, 583, 600]
s2021 = [607, 609, 580, 620]
s2022 = [714, 702, 774, 731]
s2023 = [786, 821, 807, 872]
s2024 = [935, 968, 927, 983]
s2025 = [875, 905, 0, 0]

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

plt.rc('font', **font)

x = np.arange(len(labels))  
width = 0.15

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2 * width, s2020, width, label='2020')
rects2 = ax.bar(x - width, s2021, width, label='2021')
rects3 = ax.bar(x, s2022, width, label='2022')
rects4 = ax.bar(x + width, s2023, width, label='2023')
rects5 = ax.bar(x + 2 * width, s2024, width, label='2024')
rects6 = ax.bar(x + 3 * width, s2025, width, label='2025')

ax.set_title('Количество инцидентов по кварталам')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.set_size_inches(16, 7, forward=True)

fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(aspect="equal"))

font = {'family' : 'Sans Serif',
        'weight' : 'normal',
        'size'   : 18}

data = [47, 22, 12, 4, 15]
ingredients = ['Учётные данные','Персональные данные', 'Данные платёжных карт', 'Переписка', 'Другое']


def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=16, weight="normal")

fig.set_size_inches(13, 8, forward=True)

plt.show()
```
